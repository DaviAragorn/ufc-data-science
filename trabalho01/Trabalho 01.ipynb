```python
import numpy as np
import random
```

```python
FILE_PATH = './data/column_3C.dat'
```

```python
def get_data_matrix():
    return np.loadtxt(FILE_PATH, usecols=range(6), dtype=float)
```

```python
def get_disease_mapping():
    dis = set()
    diseases = np.loadtxt(FILE_PATH, usecols=[6], dtype='U')
    for x in range(diseases.shape[0]):
        dis.add(diseases[x][0])
    disease_map = dict()
    for i, x in enumerate(dis):
        disease_map[x] = i
    return disease_map
```

```python
def get_disease_matrix():
    diseases = np.loadtxt(FILE_PATH, usecols=[6], dtype=str)
    matrix = np.zeros((310, 3))
    disease_map = get_disease_mapping()
    for x in range(matrix.shape[0]):
        matrix[x][disease_map[diseases[x][0]]] = 1
    return matrix
```

```python
def get_training_matrix(test):
    test.sort(reverse=True)
    matrix = get_data_matrix()
    for x in test:
        matrix = np.delete(matrix, x, 0)
    return matrix
```

```python
def trim_disease_matrix(matrix, test):
    test.sort(reverse=True)
    for x in test:
        matrix = np.delete(matrix, x, 0)
    return matrix
```

```python
def get_trained_matrix(test):
    X = get_training_matrix(test).T
    Y = trim_disease_matrix(get_disease_matrix(), test).T
    YXt = np.dot(Y, X.T)
    XXt = np.dot(X, X.T)
    A = np.dot(YXt, np.linalg.inv(XXt)) #TODO: XXt é a inversa
    return A
```

```python
def test():
    n_instances = get_data_matrix().shape[0]
    Yi = []
    hits = []
    class_hits = dict()
    class_used = dict()
    turn_hits = 0

    for x in get_disease_mapping().values():
        print(x)
        class_hits[x] = 0
        class_used[x] = 0

    for i in range(100):
        print(str(i)+'%')
        test = random.sample(range(0,n_instances), int(n_instances*0.2))
        A = get_trained_matrix(test)
        for x in test:
            current_disease = get_disease_matrix()[x].argmax(axis=0)
            Yi = np.dot(A, get_data_matrix()[x][:])
            class_used[current_disease] += 1
            if(Yi.argmax(axis=0) == current_disease):
                turn_hits += 1
                class_hits[current_disease] += 1
        print(turn_hits/len(test))
        hits.append(turn_hits/len(test))
        turn_hits = 0
    print(hits)
    max_hits = max(hits)
    min_hits = min(hits)
    average_hits = sum(hits)
    print('Max hits: '+str(max_hits))
    print('Min hits: '+str(min_hits))
    print('Average hits: '+str(average_hits/100))
    print('Desvio padrão: '+str(np.std(hits)))
    for x, y in zip(get_disease_mapping().values(), get_disease_mapping().keys()):
        print(y+': '+str(class_hits[x]/class_used[x]))
```
